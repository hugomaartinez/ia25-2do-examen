# Clustering


Usa esta plantilla cuando no hay columna objetivo y el ejercicio pide agrupar.

Ejemplos: segmentar clientes, agrupar pacientes, encontrar perfiles, descubrir patrones.

La pregunta clave no es “cuánto acierta”, porque no hay etiquetas. La pregunta es si los grupos son compactos, separados e interpretables.


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# CAMBIAR: usa tu archivo real.
df = pd.read_csv("data.csv")

# Si hay columnas no numéricas, quítalas o codifícalas antes.
X = df.select_dtypes(include="number")

# Revisa columnas constantes o IDs: suelen estropear clustering.
print(X.describe())
print("Columnas usadas:", list(X.columns))

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertias = []
silhouettes = []
ks = range(2, 11)

for k in ks:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    inertias.append(model.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

plt.plot(list(ks), inertias, marker="o")
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.title("Método del codo")
plt.show()

plt.plot(list(ks), silhouettes, marker="o")
plt.xlabel("Número de clusters")
plt.ylabel("Silhouette")
plt.title("Calidad de separación")
plt.show()
```

```python
FINAL_K = 3  # CAMBIAR según el codo, silhouette e interpretabilidad.

final_model = KMeans(n_clusters=FINAL_K, random_state=42, n_init=10)
df["cluster"] = final_model.fit_predict(X_scaled)

print(df["cluster"].value_counts().sort_index())
print(df.groupby("cluster")[X.columns].mean())
```

## Cómo elegir `k`

- El codo busca el punto donde la inercia deja de mejorar mucho.
- Silhouette cerca de 1 es mejor; cerca de 0 significa clusters mezclados.
- Si dos valores de `k` son parecidos, elige el que puedas explicar mejor.


## Si falla

- Si hay IDs numéricos, quítalos. No son características reales.
- Si no escalas, domina la variable con valores más grandes.
- KMeans no va bien con clusters raros o mucho ruido.
- Si los clusters no se interpretan, el ejercicio probablemente espera que expliques limitaciones.
