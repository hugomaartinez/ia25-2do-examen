# Ingeniería de Características

La **Ingeniería de Características** es el proceso de usar el conocimiento del dominio para crear nuevas características (variables) a partir de datos brutos, con el objetivo de que los algoritmos de Machine Learning funcionen mejor. Mientras que el [Análisis Exploratorio de Datos (EDA)](e2e020_eda.ipynb) se centra en *comprender* los datos, la Ingeniería de Características se centra en *transformarlos* para mejorar el rendimiento del modelo. Este suele ser uno de los pasos más impactantes en un pipeline de Machine Learning: unas características bien diseñadas pueden mejorar drásticamente la precisión del modelo, mientras que características mal elegidas pueden limitar lo que incluso los algoritmos más sofisticados pueden lograr.

## Combinación de atributos

Con frecuencia tiene sentido intentar combinar ciertas variables para obtener nuevas variables más útiles. Por ejemplo, el número total de habitaciones no es muy útil si no sabemos cuántas viviendas hay. Lo que realmente queremos es el número de habitaciones por vivienda. De manera similar, el número total de dormitorios por sí solo no es muy útil: probablemente queramos compararlo con el número de habitaciones. Y la población por vivienda también parece una combinación de atributos interesante para analizar. Creemos estas nuevas variables.

Por lo tanto, añadimos columnas con estas nuevas variables al dataframe y observamos sus correlaciones con `median_house_value`.

```
train_set["rooms_per_house"] = train_set["total_rooms"] / train_set["households"] # número de habitaciones por vivienda
train_set["bedrooms_ratio"] = train_set["total_bedrooms"] / train_set["total_rooms"] # proporción de dormitorios respecto al total de habitaciones
train_set["people_per_house"] = train_set["population"] / train_set["households"] # número de personas por vivienda
```

```
corr_matrix = train_set.corr(numeric_only=True)corr_matrix["median_house_value"].sort_values(ascending=False, key=np.abs)
```

Hemos obtenido un nuevo atributo `bedrooms_ratio` que tiene una correlación mucho mayor con `median_house_value` que el número total de habitaciones o dormitorios. Al parecer, las viviendas con una proporción más baja de dormitorios respecto al total de habitaciones tienden a ser más caras. Además, el número de habitaciones por vivienda también tiene una correlación mayor que el número total de habitaciones o dormitorios.
