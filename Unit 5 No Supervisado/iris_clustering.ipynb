# Clustering del Dataset Iris con K-Means

Vamos a aplicar el método K-Means al dataset Iris. Este dataset tiene etiquetas de clase, por lo que sería más apropiado aplicar un método de clasificación supervisada. Sin embargo, vamos a ignorar las etiquetas de clase y aplicar K-Means como clustering, comparando *a posteriori* los resultados obtenidos con las etiquetas reales.

````
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

print(iris.feature_names)
print(iris.data[:5])
print(iris.target_names)
print(iris.target[:5])
````

````
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(2,10)
for n_clusters in K:
    model = KMeans(n_clusters, n_init = 20).fit(iris.data)
    inertia.append(model.inertia_)
    
plt.plot(K, inertia, 'bx-')
plt.xticks(K)
plt.grid(True)
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
````

Con los valores predeterminados n_init='auto' e init='k-means++', [el número de iteraciones es 1](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html). Se puede observar que sin fijar la semilla, la representación en este caso varía mucho. Esto se debe a que el algoritmo K-Means es sensible a la inicialización de los centroides y no logra converger a un mínimo global en algunos casos. Al aumentar el número de iteraciones, se observa que la representación es más estable, por lo que se ha establecido n_init=100.

Una vez que el gráfico es estable, los cambios de pendiente entre 3 y 10 son bastante continuos, lo que muestra lo subjetivo que es elegir el número de clusters con este método.

````
import matplotlib.pyplot as plt
import seaborn as sns

def plot_compare_iris_n_clusters(n_clusters):
    y = KMeans(n_clusters, n_init = 20).fit_predict(iris.data) # Entrenar el modelo y predecir clusters
    
    iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # Crear un DataFrame con los datos para usar en Seaborn
    iris_df['species'] = iris.target_names[iris.target] # Agregar la especie de cada flor con su nombre
    iris_df['cluster'] = y # Agregar el cluster asignado a cada flor

    sns.scatterplot(x='petal length (cm)', y='petal width (cm)',
                    hue='cluster', # El color depende del cluster
                    palette='Set1',
                    style='species', # El estilo depende de la especie
                    data=iris_df)

    plt.xlabel('Longitud del Pétalo (cm)')
    plt.ylabel('Ancho del Pétalo (cm)')
    plt.show()
````

Ejecutando para 5 clusters:

````
plot_compare_iris_n_clusters(5)
````

````
plot_compare_iris_n_clusters(3)
````

Vemos que el resultado es bastante similar. Podríamos calcular el rendimiento del clustering con respecto a las etiquetas reales como si fuera una clasificación.

Dado que sabemos que en realidad hay 3 categorías, vamos a comparar los resultados obtenidos para 3 clusters. El número asignado a cada cluster no necesariamente coincide con el número asignado a cada clase real, por lo que para compararlos, asignaremos a cada cluster el número de clase real que aparece con mayor frecuencia en él.

````
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42).fit(iris.data)
df = pd.DataFrame({'cluster': kmeans.labels_, 'real_class': iris.target_names[iris.target]})
df
````

Calculamos la moda de cada cluster y asignamos a cada cluster el número de clase real que aparece con mayor frecuencia en él.

````
modes = df.groupby('cluster')['real_class'].agg(pd.Series.mode)
modes
````

Y añadimos la columna de la moda de cada cluster a nuestro DataFrame (asumiendo que hay una sola moda clara en cada cluster). Sería necesario ser más preciso si hubiera empates, y sería problemático si dos clusters tuvieran la misma moda.

````
df['cluster_mode'] = df['cluster'].map(modes)
df
````

Y ahora que el mismo número de cluster coincide con el mismo número de clase real, podemos compararlos como si fuera una clasificación.

````
from sklearn.metrics import classification_report

y_pred = KMeans(3, n_init = 20).fit_predict(iris.data)
print(classification_report(df.real_class, df.cluster_mode))
````
