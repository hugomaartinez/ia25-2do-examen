# Preparar el examen sin internet


Esta notebook es para ejecutarla al principio del examen, cuando todavía hay internet.

Objetivo: dejar datasets y modelos en carpetas locales para que el resto del examen no dependa de Kaggle, Hugging Face, Ultralytics ni APIs.


## Regla importante

Con Kaggle hay dos opciones:

1. Confiar en la caché de `kagglehub`.
2. Descargar el dataset y copiar el CSV a una ruta local controlada.

Para examen, usa la opción 2. La caché puede funcionar, pero es menos visible y más fácil de romper si cambia el entorno.


````python
from pathlib import Path
import shutil

DATASETS_DIR = Path("datasets")
DATASETS_DIR.mkdir(exist_ok=True)

print("Carpeta local de datasets:", DATASETS_DIR.resolve())
````

````python
# Descargar King County desde Kaggle y dejar CSV local.
# Ejecutar solo mientras haya internet.

from pathlib import Path
import kagglehub
import shutil

king_path = Path(kagglehub.dataset_download("harlfoxem/housesalesprediction"))
king_csv = king_path / "kc_house_data.csv"

out_dir = Path("datasets/king-county")
out_dir.mkdir(parents=True, exist_ok=True)
shutil.copy2(king_csv, out_dir / "kc_house_data.csv")

print("King County guardado en:", (out_dir / "kc_house_data.csv").resolve())
````

````python
# Descargar Thyroid desde Kaggle y dejar CSV local.
# Ejecutar solo mientras haya internet.

from pathlib import Path
import kagglehub
import shutil

thyroid_path = Path(kagglehub.dataset_download("emmanuelfwerr/thyroid-disease-data"))
thyroid_csv = thyroid_path / "thyroidDF.csv"

out_dir = Path("datasets/thyroid")
out_dir.mkdir(parents=True, exist_ok=True)
shutil.copy2(thyroid_csv, out_dir / "thyroidDF.csv")

print("Thyroid guardado en:", (out_dir / "thyroidDF.csv").resolve())
````

````python
# Descargar pesos YOLO si vas a usar computer-vision/yolo.
# Ejecutar solo mientras haya internet.

from pathlib import Path
from ultralytics import YOLO

models_dir = Path("models")
models_dir.mkdir(exist_ok=True)

# Ultralytics descarga si no existe. Después puedes copiar el .pt si hace falta.
model = YOLO("yolov8n.pt")
print("YOLO cargado. Si no estaba descargado, queda cacheado por Ultralytics.")
````

````python
# Descargar CLIP/SentenceTransformer si vas a usar embeddings de imagen.
# Ejecutar solo mientras haya internet.

from sentence_transformers import SentenceTransformer

clip_model = SentenceTransformer("clip-ViT-B-32")
print("CLIP descargado/cacheado correctamente")
````

````python
# Comprobación offline rápida.
# Después de cortar internet, esta celda debería seguir encontrando los CSV locales.

from pathlib import Path

required_files = [
    Path("datasets/king-county/kc_house_data.csv"),
    Path("datasets/thyroid/thyroidDF.csv"),
]

for file_path in required_files:
    print(file_path, "OK" if file_path.exists() else "FALTA")
````
